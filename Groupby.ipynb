```python
import pandas as pd
```

```python
df_astop = pd.read_csv('/content/Aspirantes TOP - Limpio - Todo.csv', sep=',',header=0)

df_astop.head(2)
```

```python
df_astop.groupby('Género').count()
```

```python
df_astop.groupby('Género').sum()
```

```python
df_astop.groupby('Género').max()
```

```python
df_astop.groupby('Género').mean()
```

```python
df_astop.groupby('Nivel').mean().loc[3]
```

```python
df_astop.groupby('Género').sum().reset_index()
```

```python
df_astop.groupby('Género').agg(['min', 'max'])
```

```python
df_astop.groupby('Género').agg({'Edad':['min','max'],'Misiones':'sum'})
```

```python
df_astop.groupby(['Género','Nivel']).count()
```
