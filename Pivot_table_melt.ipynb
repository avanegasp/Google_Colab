```python
import pandas as pd
import numpy as np
```

```python
c9_astop = pd.read_csv('/content/C9 - Curso Aspirante y Admitidos - 25 de abril - toda.csv',sep=',',header=0)
c9_astop.head(2)
```

```python
c9_astop.pivot_table(index='first_name',columns='género',values='Nivel')
```

```python
c9_astop.pivot_table(index='first_name',columns='género',values='Nivel',aggfunc='mean')
```
